# Ploting By Python

```python
# import libraries 
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

### First we import data set form seaborn library

```python
#import data set
kashti = sns.load_dataset('titanic')
kashti
```

### 2. We check the info of data set

```python
# first we check the info of data set
kashti.info()
```

### 3. Check the Null (NAN) Values of Data Set

```python
# second we check the null values in data set
kashti.isnull()
```

```python
kashti.isnull().sum()
```

## Now We visualize the data with the help of Plots

### 1.Bar Plot

```python
# make a plot
kashti.survived.value_counts().plot(kind='bar')
```

### 2.Seaborn count Bar Plot

```python
sns.countplot(x="survived", hue="sex" ,data=kashti)
```

### 3.Matplotlib Scatter plot

```python
#scatter plot also remeber we dont use hue in plt scatter we use c for colours
plt.scatter(data=kashti, x="age",y="fare", c="pclass")
```

```python
# scatter plot in sns and here we use hue 
sns.scatterplot(data=kashti, x="age", y="fare", hue="pclass")
```

### 4.Seaborn Bar Plot

```python
# comparison rent on sex values
sns.barplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
kashti.columns
```

### 5.Seaborn Pair Plot

```python
sns.pairplot(kashti)
#in this we can create a plot in which every column is a variable and every row is a observation
```

```python
# to create a hist plot for numeric values if we see a curve in graphs then data is correct
kashti.hist()
```

```python
kashti.hist(bins=10 , figsize=(9,7))
```

```python
# for removing grid we use grid 
kashti.hist(bins=10 , figsize=(9,7) , grid=False)
```

### 6.Seaborn Box Plot

```python
# create box plot
sns.boxplot(data=kashti, x="sex", y="fare", hue="class")
```

### 7.Seaborn Box-en-Plot

```python
# we change the x n y axis title
sns.boxenplot(data=kashti, x="sex", y="fare", hue="class")
```

```python
#facitgrid plot
```

### 8.Seaborn Strip Plot

```python
sns.stripplot(data=kashti, x="sex", y="fare", hue="class")
```

### 9.Matplotlib Heatmap Plot

```python
# here we use typecasting to remove errors
# corr is the relation between two variables

numeric_columns = kashti.select_dtypes(include=['float64', 'int64']).columns 
corr_matrix = kashti[numeric_columns].corr() 
plt.figure(figsize=(10, 10)) 
sns.heatmap(corr_matrix, annot=True) 
plt.title('Correlation Heatmap')
plt.show()
```

### 10.Seaborn violin Plot

```python
#violant plot
sns.violinplot(data=kashti, x="sex", y="fare", hue="class")
```
